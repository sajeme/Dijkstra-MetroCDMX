```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

Se lee el archivo excel, el directorio se busca con ayuda de la libreria OS

```python
directory = os.getcwd()
df = pd.read_excel(directory+'\EstacionesMetroCDMX.xlsx')
```

```python
df.head()
```

```python
    tren = nx.from_pandas_edgelist(df,source='Salida',target='Llegadas',edge_attr='Distancia')
```

Los nodos que serian las estaciones de tren

```python
tren.nodes()
```

```python
tren.edges()
```

```python
tren.order()
```

centralidad desordenada y ordenada de mayor a menor

```python
print(nx.degree_centrality(tren))
```

```python
most_influential = nx.degree_centrality(tren)
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

Mapa de grafo usando la libraria matplotlib

```python
plt.figure(figsize=(50,50))
nx.draw_networkx(tren)
plt.show()
```

calculo de distancia mas corta y longitud con el algoritmo dijkstra

```python
djk_path= nx.dijkstra_path(tren, source='El Rosario', target='Pino Suárez', weight='Distancia')
djk_path
```

```python
len(djk_path)
```
