```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Se lee el archivo excelcsv

```python
df = pd.read_excel('C:/Users/Personal/TAREA/EstacionesMetroCDMX.xlsx')
```

```python
df.head()
```

```python
    tren = nx.from_pandas_edgelist(df,source='Salida',target='Llegadas',edge_attr='Distancia')
```

Solo para ver nodo

```python
tren.nodes()
```

centralidad desordenada y ordenada

```python
print(nx.degree_centrality(tren))
```

```python
most_influential = nx.degree_centrality(tren)
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

edges y order

```python
tren.edges()
```

```python
tren.order()
```

mapa de grafo

```python
plt.figure(figsize=(50,50))
nx.draw_networkx(tren)
plt.show()
```

calculo de distancia mas corta y longitud con dijkstra

```python
djk_path= nx.dijkstra_path(tren, source='El Rosario', target='Pino Suárez', weight='Distancia')
djk_path
```
